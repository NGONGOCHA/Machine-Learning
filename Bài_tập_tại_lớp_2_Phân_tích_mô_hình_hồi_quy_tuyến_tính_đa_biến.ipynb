```python
import numpy as np

X = [[1, 2, 3, 4, 5],
     [6, 7, 8, 9, 2],
     [4, 5, 7, 2, 1]]
print("ma trận X: \n", np.array(X))
X = np.reshape(X, (5,3))
print("chuyển ma trận X sang kích thước (5,3): \n", X)
```

image.png

```python
w = [2,4,2]
w = np.reshape(w, (3,1))
print(w)
```

image.png

```python
hwx = np.dot(X, w)
print("hw(x): \n", hwx)
print("chiều: \n", hwx.ndim)
print("kích thước: \n", hwx.shape)
```

image.png

```python
y= np.reshape([1,2, 3, 2, 1], (5, 1))
m = y.shape[0]
n = np.square(hwx-y)
J = (1/(2*m))*np.sum(n)
print("J(w) = " ,J)
```

image.png

```python
XT = np.transpose(X)
print("ma tran ban dau:\n",X)
print("ma tran chuyen vi:\n",XT)

dh_J = (1/(m))*np.sum(XT.dot(np.square(hwx-y)))
print("cong thuc tinh dao ham rieng theo dang ma tran:",dh_J)
```
